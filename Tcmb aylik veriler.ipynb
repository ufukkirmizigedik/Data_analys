```python
import pandas as pd
import matplotlib.pyplot as py
```

```python
df = pd.read_csv('tcmb_aylik_veriler.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Tarih'] = pd.to_datetime(df['Tarih'],format='%Y-%m')
```

```python
print(df['Tarih'].dtype)
```

```python
yearly_data = df.groupby(df['Tarih'].dt.year)['tuketici_kredisi_faizi'].sum()
```

```python
yearly_data.plot.area(figsize=(12,5))
```

```python
yillik_konut_satis = df.groupby(df['Tarih'].dt.year)['toplam_ikinciel_konut_satisi'].sum()
```

```python
yillik_konut_satis.plot.area(figsize=(12,5))
```

```python
yillik_sifir_konut = df.groupby(df['Tarih'].dt.year)['toplam_sifir_konut_satisi'].sum()
```

```python
yillik_sifir_konut.plot.area(figsize=(12,5))
```

```python
df['yil'] = df['Tarih'].dt.year
```
